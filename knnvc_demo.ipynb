```python
import IPython.display as ipd
from IPython.display import Audio
import torch, torchaudio
import os
```

```python
knn_vc = torch.hub.load('bshall/knn-vc', 'knn_vc', prematched=True, trust_repo=True, pretrained=True, device='cuda')
```

```python
# from google.colab import drive
# drive.mount('/content/drive/')
```

```python
base_dir = "/media/shashank/data1/iisc/projects/capstone_project/templates/assets/audio_files/"
source = 228
target=225

tar_path = f"{base_dir}p{str(target)}/"
ref_wav_paths = [os.path.join(tar_path, file) for file in os.listdir(tar_path)]
src_wav_path = f"{base_dir}p{str(source)}/p{str(source)}_003.wav"

print(src_wav_path)
print(ref_wav_paths[0])

results_dir = "/media/shashank/data1/iisc/projects/capstone_project/templates/assets/audio_files/knn_results/"
```

```python
# path to 16kHz, single-channel, source waveform
# src_wav_path = '/content/drive/MyDrive/capstone_G7/sample_40/all_samples/p225_003.wav'
# list of paths to all reference waveforms (each must be 16kHz, single-channel) from the target speaker
# ref_wav_paths = ['/content/drive/MyDrive/capstone_G7/sample_40/all_samples/p228_003.wav', ]

query_seq = knn_vc.get_features(src_wav_path)
matching_set = knn_vc.get_matching_set(ref_wav_paths)
```

```python
out_wav = knn_vc.match(query_seq, matching_set, topk=4)
```

```python
out_file_path = f"{results_dir}p{source}xp{target}.wav"
torchaudio.save(out_file_path, out_wav[None], 16000)
```
